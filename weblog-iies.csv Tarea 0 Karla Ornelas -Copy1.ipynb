Principalmente se limpiara el dataset para que se puedan obtener resultados del mismo , se utilizaran librerias de panda y seaborn. 

El código es:


```python
#3 Código y ejemplo de la solución

#Librerias utilizadas
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

#Código de la solución
duplicate = pd.read_csv('weblog-iies.csv')

duplicate.dtypes
data=pd.DataFrame(duplicate)
data['Time']=data['Time'].map(lambda x: x.lstrip('['))
from datetime import timedelta
data['Time'] = data['Time'].map(lambda x: timedelta(hours=int(x.split(':')[0]), minutes=int(x.split(':')[1].split('.')[0]), seconds=int(x.split(':')[1].split('.')[1])))
data = data.rename(columns={'packets': 'Packets'})
data['Address'] = data['Address'].map(lambda x: str(x).lstrip('0') if pd.notnull(x) else x)

data.describe()
data.isnull().sum().sum()
data = data.dropna()
data.isnull().sum().sum()

data['Time'].iloc[0]
data['Time_str'] = data['Time'].astype(str)
data['month'] = data['Time_str'].str.slice(3, 5)
data['day'] = data['Time_str'].str.slice(0, 2)

#Funciones
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
data['Type'].value_counts().head(40).plot.bar(color = color)
plt.title('Tipos mas populares usados por los usuarios', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
data['URL'].value_counts().head(40).plot.bar(color = color)
plt.title('URL mas populares entre los usuarios', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.copper(np.linspace(0, 1, 40))
data['Time'].value_counts().head(40).plot.bar(color = 'cyan')
plt.title('Usuarios Con mas tiempo en la pagina', fontsize = 20)
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

color = plt.cm.Wistia(np.linspace(0, 1, 40))
data['Packets'].value_counts().head(40).plot.bar(color = 'seagreen')
plt.title('Packets mas populares', fontsize = 20)
```
