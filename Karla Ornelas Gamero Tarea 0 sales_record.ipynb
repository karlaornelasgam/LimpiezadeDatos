```python
import matplotlib.pyplot as plt
#hacer un np.array y graficar las medias

import numpy as np
import matplotlib.pyplot as plt

def cal_min_max_prom(datosfm):
    medias = []
    mins = []
    maxs = []
    
    for index in datosfm:
        valores = datosfm[index].describe()
        prom = valores.iloc[1]
        minimo = valores.iloc[2]
        maximo = valores.iloc[3]
        medias.append(prom)
        mins.append(minimo)
        maxs.append(maximo)
    
    return medias, mins, maxs

fmtemporal = temp_data[temp_data.columns[3:]]

medias, mins, maxs = cal_min_max_prom(fmtemporal)

print(fmtemporal.describe())
print(medias)
print(mins)
print(maxs)

plt.plot(medias)
plt.plot(mins)
plt.plot(maxs)
plt.show()
```

```python
datosl =  temp_data.copy()
media = datosl["Total Revenue"].describe().iloc[1]
datosl = datosl.fillna(value=media)
```

```python
#limpiar renglos para s1--s31 con valor 0 
j=0
for i in datosl.itertuples():
    promedio = np.array(i[12:]).mean()
    minimo = np.array(i[12:]).min()
    maximo = np.array(i[12:]).max()
    if promedio == 0:
        print("Eliminado: ",j)
        datosl = datosl.drop(i.Index, axis=0)
    j=j+1
datosl
```
